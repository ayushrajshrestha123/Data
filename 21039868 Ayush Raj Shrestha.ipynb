```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.DataFrame()
for each in os.listdir('C:\\Users\\XPS1\\Downloads\\csv_files_cw'):
    files = pd.read_csv(each)
    df = pd.concat([df, files])
df.to_csv('New.csv')
df
```

```python
df=df.dropna() #removes the rows that contains NULL values.
df=df.dropna(axis=0)
df
```

# Write a python program to convert Quantity Ordered and Price Each to numeric.

```python
df["Quantity Ordered"]=pd.to_numeric(df["Quantity Ordered"]) #convert to numeric
df["Price Each"]=pd.to_numeric(df["Price Each"])
df.info()
```

# Create a new column named Month from Ordered Date of updated dataframe and convert it to integer as data type.

```python
df=df.assign(Month=df["Order Date"])
df["Month"] = df["Order Date"].str.split('/').str[0] #splitting table and only keeping the first part of the date which is month
df["Month"] = df["Month"].astype('int')#sets the data type of the column to int
df.info()
df
```

# Create a new column named City from Purchase Address based on the value in updated dataframe.

```python

df['City'] = df['Purchase Address'].str.split(',').str[1] # splitting the purchase address's 2nd part, which is city
df
```

# Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.

```python
Sum=df["Quantity Ordered"].sum()
print(Sum) # prints sum
mean=df["Quantity Ordered"].mean()
print("The mean is",mean)# prints mean
standardDeviation=df["Quantity Ordered"].std()
print("The std is",standardDeviation)# prints standard deviation
skewness=df["Quantity Ordered"].skew()
print("The skewness is",skewness)# prints skewness
kurtosis=df["Quantity Ordered"].kurtosis()
print("The kurtosis is",kurtosis)# prints kurtosis
```

# Which Month has the best sales? and how much was the earning in that month? Make a bar graph of sales as well.

```python
df["Price"]=df["Quantity Ordered"]*df["Price Each"] #creates new column named "price" for further calculation
sum_price = df.groupby('Month')['Price'].sum() #sum of price in a month
max_price = sum_price.idxmax()
max_earning = sum_price[max_price]
print("The month with the best sales is", max_price) 
print("The earning in that month was", max_earning)

months=range(1,13)
plt.figure(figsize=(15,6))
plt.bar(months,sum_price,color="Red")
plt.xlabel("Months")
plt.ylabel("Total Sales")
plt.title("Total Sales in all Months")
plt.show()=
```

# Which city has sold the highest product?

```python
plt.figure(figsize=(15,6))
product_sold=[] #to hold the products sold, an empty list is made
Cities=df["City"].unique() #to prevent redundancy, unique function is called
for i in range(0,len(Cities)):
    City=(Cities[i])
    TotalProductSold=sum(df[df["City"]==City]["Quantity Ordered"])
    product_sold.append(TotalProductSold)
plt.bar(Cities,product_sold,color="Blue")
plt.xlabel("Cities")
plt.ylabel("Total Product Sold")
plt.title("Total Product sold in Cities")
plt.show()
```

# Which product was sold the most in overall? Illustrate it through bar graph.

```python
plt.figure(figsize=(50,20))
plt.xlabel("Products")
plt.ylabel("Total Products Sold")
plt.title("Highest Product Sold")
Products=df["Product"].unique()
all_products=[]
for each in Products:
    TotalQuantity=sum(df[df["Product"]==each]["Quantity Ordered"])
    all_products.append(TotalQuantity)
plt.bar(Products,all_products,color="green")
plt.show()
```
